```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
```

```python
diabetes = datasets.load_diabetes()
print(diabetes.keys())

# ['data', 'target', 'frame', 'DESCR', 'feature_names', 'data_filename', 'target_filename'])


print(diabetes.data)
print(diabetes.target)
print(diabetes.DESCR)
print(diabetes.feature_names)
print(diabetes.data_filename)
print(diabetes.target_filename)


# Features: ['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'] scaled between -1 to 1 such that sum of squares is 1.
```

```python
diabetes_X = diabetes.data[:, np.newaxis, 4] # taking only one feature for better visualisation, for multiple features, just replace it by diabetes.data

# print(diabetes_X)

diabetes_X_train = diabetes_X[:-30]
diabetes_X_test = diabetes_X[-30:]

diabetes_Y = diabetes.target

diabetes_Y_train = diabetes.target[:-30]
diabetes_Y_test = diabetes.target[-30:]
```

```python
model = linear_model.LinearRegression()

model.fit(diabetes_X_train, diabetes_Y_train)

prediction = model.predict(diabetes_X_test)

error = mean_squared_error(diabetes_Y_test, prediction)

print("Mean Square Error: ", error)
print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)
```
